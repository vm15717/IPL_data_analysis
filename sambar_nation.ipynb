```python
df_pred2 = df_pred.drop(['Match'], axis = 1)
df_pred3 = df_pred2.drop(['Winner'], axis = 1)
```

```python
from sklearn.metrics import cohen_kappa_score
import pandas as pd

# Initialize an empty matrix
people = df_pred3.columns
kappa_matrix = pd.DataFrame(index=people, columns=people, dtype=float)

# Calculate Cohen's Kappa for each pair
for p1 in people:
    for p2 in people:
        if p1 == p2:
            kappa_matrix.loc[p1, p2] = 1.0  # Perfect agreement with self
        else:
            # Filter out rows where either p1 or p2 is 'NaN' (as a string)
            mask = (df_pred3[p1] != 'NaN') & (df_pred3[p2] != 'NaN')
            y1 = df_pred3[p1][mask]
            y2 = df_pred3[p2][mask]

            # Check if there's enough data left
            if len(y1) > 0:
                kappa_matrix.loc[p1, p2] = cohen_kappa_score(y1, y2)
            else:
                kappa_matrix.loc[p1, p2] = None  # or np.nan, if preferred

print(kappa_matrix)
```
