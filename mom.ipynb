```python
import json
```

```python
from os import listdir
from os.path import isfile, join
from collections import Counter
mypath = './ipl_json/'
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]
```

```python
onlyfiles.remove('README.txt')
len(onlyfiles)
```

```python

mom = {}
for f in onlyfiles:
    with open(mypath + f, 'r') as file:
        data = json.load(file)
        try:
            mom_f = data['info']['player_of_match'][0]
        except:
            print(f"Match No {data['info']['event']['match_number']}: {data['info']['outcome']['result']}")
        if mom_f in mom.keys():
            mom[mom_f] += 1
        else:
            mom[mom_f] = 1
```

```python
for key, value in sorted(mom.items(), key=lambda x: x[1], reverse=True):
    print(key, value)
```

```python
def extract_player_data(allfiles, player_name):
    # Load JSON data
    for f in allfiles:
        with open(mypath + f, 'r') as file:
            data = json.load(file)
    
        player_data = {
            'batting': [],
            'bowling': [],
            'dismissals': []
        }
        
        # Iterate through innings and deliveries
        for inning in data.get('innings', []):
            team_name = inning.get('team')
            for over_data in inning.get('overs', []):
                over_number = over_data.get('over')
                count = 0
                for delivery in over_data.get('deliveries', []):
                    # Batting Data
                    count += 1
                    if delivery.get('batter') == player_name:
                        player_data['batting'].append({
                            'over': over_number,
                            'ball_number': count,
                            'bowler': delivery.get('bowler'),
                            'runs': delivery.get('runs', {}).get('batter', 0),
                            'total_runs': delivery.get('runs', {}).get('total', 0)
                        })
                    
                    # Bowling Data
                    if delivery.get('bowler') == player_name:
                        player_data['bowling'].append({
                            'over': over_number,
                            'ball_number': count,
                            'batter': delivery.get('batter'),
                            'runs': delivery.get('runs', {}).get('total', 0),
                            'extras': delivery.get('runs', {}).get('extras', 0)
                        })
                    
                    # Dismissal Data
                    for wicket in delivery.get('wickets', []):
                        if wicket.get('player_out') == player_name or player_name in [f.get('name') for f in wicket.get('fielders', []) if f.get('name')]:
                            player_data['dismissals'].append({
                                'over': over_number,
                                'ball_number': count,
                                'bowler': delivery.get('bowler'),
                                'kind': wicket.get('kind'),
                                'player_out': wicket.get('player_out')
                            })
        
        return player_data


player_name = 'TM Head'#input('Enter player name: ')
result = extract_player_data(onlyfiles, player_name=player_name)
print(json.dumps(result, indent=2))
```
